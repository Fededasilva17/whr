```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("2019.csv")
df.info()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle("Histograms of Score, GDP per capita, Healthy life expectancy, and Freedom to make life choices", fontsize=16)

# Histogramas con colores, bordes y transparencia
# Score
axes[0, 0].hist(df['Score'], bins=10, color='skyblue', edgecolor='black', alpha=0.7)
axes[0, 0].set_title('Score')
axes[0, 0].set_xlabel('Score')
axes[0, 0].set_ylabel('Frequency')

# GDP per capita
axes[0, 1].hist(df['GDP per capita'], bins=10, color='salmon', edgecolor='black', alpha=0.7)
axes[0, 1].set_title('GDP per capita')
axes[0, 1].set_xlabel('GDP per capita')
axes[0, 1].set_ylabel('Frequency')

# Healthy life expectancy
axes[1, 0].hist(df['Healthy life expectancy'], bins=10, color='lightgreen', edgecolor='black', alpha=0.7)
axes[1, 0].set_title('Healthy life expectancy')
axes[1, 0].set_xlabel('Healthy life expectancy')
axes[1, 0].set_ylabel('Frequency')

# Freedom to make life choices
axes[1, 1].hist(df['Freedom to make life choices'], bins=10, color='violet', edgecolor='black', alpha=0.7)
axes[1, 1].set_title('Freedom to make life choices')
axes[1, 1].set_xlabel('Freedom to make life choices')
axes[1, 1].set_ylabel('Frequency')

# Ajustar el diseo y mostrar
plt.tight_layout(rect=[0, 0, 1, 0.96])  # rect ajusta el espacio para el ttulo principal
plt.show()
```

```python
# Configuracin general para los grficos
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle("Boxplots of Score, GDP per capita, Healthy life expectancy, and Freedom to make life choices", fontsize=16)

# Boxplot para 'Score'
axes[0, 0].boxplot(df['Score'], patch_artist=True, boxprops=dict(facecolor='skyblue', color='black'), 
                   medianprops=dict(color='red'), whiskerprops=dict(color='blue'))
axes[0, 0].set_title('Score')
axes[0, 0].set_ylabel('Score')

# Boxplot para 'GDP per capita'
axes[0, 1].boxplot(df['GDP per capita'], patch_artist=True, boxprops=dict(facecolor='salmon', color='black'), 
                   medianprops=dict(color='green'), whiskerprops=dict(color='purple'))
axes[0, 1].set_title('GDP per capita')
axes[0, 1].set_ylabel('GDP per capita')

# Boxplot para 'Healthy life expectancy'
axes[1, 0].boxplot(df['Healthy life expectancy'], patch_artist=True, boxprops=dict(facecolor='lightgreen', color='black'), 
                   medianprops=dict(color='orange'), whiskerprops=dict(color='gray'))
axes[1, 0].set_title('Healthy life expectancy')
axes[1, 0].set_ylabel('Healthy life expectancy')

# Boxplot para 'Freedom to make life choices'
axes[1, 1].boxplot(df['Freedom to make life choices'], patch_artist=True, boxprops=dict(facecolor='violet', color='black'), 
                   medianprops=dict(color='blue'), whiskerprops=dict(color='brown'))
axes[1, 1].set_title('Freedom to make life choices')
axes[1, 1].set_ylabel('Freedom to make life choices')

# Ajuste del diseo y visualizacin
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

```python
# Configuracin para la visualizacin
fig, axes = plt.subplots(4, 2, figsize=(14, 16), gridspec_kw={'height_ratios': [3, 1, 3, 1]})
fig.suptitle("Histograms and Boxplots of Variables", fontsize=18)

variables = ['Score', 'GDP per capita', 'Healthy life expectancy', 'Freedom to make life choices']
colors = ['skyblue', 'salmon', 'lightgreen', 'violet']

for i, var in enumerate(variables):
    # Histograma en la columna de la izquierda para cada variable
    axes[i, 0].hist(df[var], bins=8, color=colors[i], edgecolor='black', alpha=0.7)
    axes[i, 0].set_title(f'Histogram of {var}')
    axes[i, 0].set_ylabel('Frequency')
    
    # Boxplot en la columna de la derecha para cada variable
    axes[i, 1].boxplot(df[var], vert=False, patch_artist=True,
                       boxprops=dict(facecolor=colors[i], color='black'),
                       medianprops=dict(color='red'))
    axes[i, 1].set_title(f'Boxplot of {var}')
    axes[i, 1].set_yticks([])  # Ocultar marcas del eje Y en el boxplot

# Ajuste del diseo y visualizacin
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
